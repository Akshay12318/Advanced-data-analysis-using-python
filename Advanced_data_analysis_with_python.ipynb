```python
# discrete data-unique values countable-useful for continuous data

col_list=["Region","Quantity","Segment","Ship Mode"]
for i in col_list:
    plt.figure(figsize=(10,10))
    sns.countplot(data=df,x=i)
    plt.title(f"{i} countplot")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# When two beams/bars are attached then it is histogram, else it is barplot
# histogram is useful when the values is continuous (uncountable - infinity)
col_list=["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(10,10))
    plt.hist(data=df,x=i)
    plt.title(f"{i} countplot")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#distribution plot- it was made on continuous data
col_list=["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(10,10))
    sns.displot(data=df,x=i)
    plt.title(f"{i} Distribution plot")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#distribution plot- it was made on continuous data
col_list=["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(10,10))
    sns.boxplot(data=df,x=i)
    plt.title(f"{i} Box plot")
    plt.xlabel(i)
    plt.show()
```

```python
# kde-kernal density estimation plot
col_list=["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(10,10))
    sns.kdeplot(data=df,x=i)
    plt.title(f"{i} KDE Plot")
    plt.xlabel(i)
```

```python
Col_list=["Region","Quantity","Segment","Ship Mode"]
for i in Col_list:
    plt.figure(figsize=(5,5))
    plt.pie(df[i].value_counts(),labels=df[i].value_counts().index,autopct="%.1f%%")
    plt.title(f"{i} Pie Chart")
    plt.show()
```

```python
#violin plot
col_list=["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(5,5))
    sns.violinplot(data=df,x=i)
    plt.title(f"{i} Violin Plot")
    plt.xlabel(i)
    plt.show()
```
